### Pertanyaan Analisa
Bagaimana karakteristik transaksi pelanggan member di cabang B untuk produk Sports and Travel?

```python
cabang_b_member_sports = df[((df['Branch'] == 'B')& (df['Customer type']=='Member'))&(df['Product line']=='Sports and travel')]
```

### Pertanyaan Analisa
Bagaimana distribusi total penjualan per bulan dan bulan mana yang memiliki total penjualan?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
```

```python
penjualan_bulanan = df.groupby('Month')['Total'].sum().reset_index()
penjualan_bulanan_sorted = penjualan_bulanan.sort_values(by='Total', ascending=False)
```

```python
penjualan_bulanan_sorted
```

### ðŸ’¡Insight
Berdasarkan hasil sorting, transaksi dengan nilai total tertinggi berasal dari bulan Januari, dan diikuti pada bulan Maret, setelahnya pada bulan Februari.

### Pertanyaan Analisa
Bagaimana total penjualan dimasing-masing cabang?

```python
penjualan_cabang = df.groupby('Branch').agg({'Total': 'sum',
    'Quantity': 'sum',
    'gross income': 'sum'}).reset_index()

penjualan_cabang
```

### ðŸ’¡Insight
Berdasarkan hasil agregasi, terlihat bahwa cabang C memiliki total penjualan, jumlah barang terjual, dan gross income tertinggi dibandingkan cabang A dan B. Hal ini menunjukkan bahwa performa penjualan cabang C lebih unggul secara keseluruhan.

### Pertanyaan Analisa
Bagaimana perbandingan total penjualan berdasarkan cabang dan gender?

```python
tabel_pivot = pd.pivot_table(df, values = 'Total',index = 'Branch',columns='Gender',aggfunc='sum')
tabel_pivot
```

### ðŸ’¡Insight
Dari tabel pivot terlihat bahwa di cabang A, total penjualan antara pelanggan perempuan dan laki-laki relatif seimbang. Di cabang B, total penjualan dari pelanggan laki-laki sedikit lebih tinggi dibandingkan perempuan. Sementara itu, di cabang C pelanggan perempuan memberikan kontribusi penjualan yang lebih besar dibandingkan customer laki-laki


### Pertanyaan Analisa
Bagaimana distribusi customer type pada setiap cabang?

```python
crosstab = pd.crosstab(df['Customer type'],df['Branch'])
crosstab
```

### ðŸ’¡Insight
Pada tabel crosstab terlihat bahwa pada cabang A & cabang B customer normal lebih dominan daripada customer Member. Dan pada cabang C customer Member lebih dominan daripada customer biasa.

### Pertanyaan Analisa
Pada hari apa transaksi penjualan paling banyak terjadi?

```python
df.info()
```

```python
# mengubah struktur data pada date
df["Date"] = pd.to_datetime(df["Date"],format = "%m/%d/%Y")
```

```python
df.info()
```

```python
# mengambil nama hari dari tanggal transaksi
df['Day'] = df['Date'].dt.day_name()
```

```python
# hitung jumlah transaksi per hari
transaksi_per_hari = df['Day'].value_counts()
transaksi_per_hari
```

### ðŸ’¡Insight
Dari hasil eksplorasi berdasarkan hari, terlihat bahwa jumlah transaksi paling banyak terjadi pada hari Sabtu dibandingkan hari lainnya. Hal ini menunjukkan adanya pola waktu dalam aktivitas transaksi penjualan.
